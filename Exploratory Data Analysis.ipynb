This Dataset contain list of 15 different classes of images with there image count.

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

files = os.listdir("PlantVillage")
file_classes = []
for f in files:
    file_classes.append("PlantVillage/"+f)

labels = []
counts = []
for f,images in zip(files,file_classes):
    labels.append(f)
    counts.append(len(os.listdir(images)))
#     print(f+": "+str(len(os.listdir(images))))

plt.barh(labels,counts)
plt.title("Dataset Description")
plt.xlabel("Image Count")
plt.ylabel("Image Classes")
plt.rcParams["figure.figsize"] = [10,5]
plt.show()
```

From above Graph, we can see Tomato_Tomato_YellowLeaf_curl_virus has highest Number of images with count of more than 3000 and Potato__healthy have less Number of Image with count of less than equal to 100. we also get to know that our will give good results in detecting Tomato_Tomato_YellowLeaf_curl_virus class type leaf images.

When we explore the dataset, we get to know that ther 3 major classes which our model is used, They are list down below:

1. Tomato
2. Potato
3. Pepper

```python
import networkx as nx
import re
```

```python
Major_classes = ["Tomato","Pepper","Potato"]
class_dict = {}

for cls in Major_classes:
    regx = r'\b'+str(cls)+'.*?\\b'
    cls_lst = []
    for x in labels:
        if(len(re.findall(regx,x))>0):
            cls_lst.append(x)
    class_dict[cls] = cls_lst
    
#find node
nodes = []
nodes.append('label')
for x in Major_classes:
    nodes.append(x)
for x in labels:
    nodes.append(x)

#finding edge
edges = []
for x in Major_classes:
    edges.append(('label',x))
for x in class_dict:
    for y in class_dict[x]:
        edges.append((x,y))
```

# Graphical Relationship between Classes

```python
G=nx.Graph()

for nd in nodes:
    G.add_node(nd,label = nd)
for x,y in edges:
    G.add_edge(x,y)
nx.draw(G,with_labels = True)
plt.show()
```

From above graph, We can get to know about relationship within dataset and also get to know about hierarchy of classes
